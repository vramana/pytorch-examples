```python
import numpy as np
import matplotlib.pyplot as plt

eta_max = 0.1
eta_min = 0.001
T_max = 100

t = np.arange(T_max)
lrs = eta_min + 0.5 * (eta_max - eta_min) * (1 + np.cos(np.pi * t / T_max))

plt.plot(t, lrs)
plt.xlabel("Epoch")
plt.ylabel("Learning Rate")
plt.title("Cosine Annealing (Formula)")
plt.grid(True)
plt.show()
```
